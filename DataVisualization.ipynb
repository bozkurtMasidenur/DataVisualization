```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(rc={'figure.figsize':(5,5)})
```

```python
# dataset okuma - dataset yolu verilerek okunur
# CSV (comma seperated values)
df= pd.read_csv("/content/world-happiness-report.csv")
df.head(10).T # T ile transpoz alınır
```

```python
df.info() # dataset hakkında bilgi verir
```

```python
df['Country name'].value_counts() # country name de bulunan değişkenlerin kaç kere tekrar ettiğinin sayısı
```

```python
df.describe().T 
# describe ile dataset kolonlarının aggregation fonksiyonları sonuçlarını görebiliriz
```

```python
from os import mkdir # make directory - siyah ekran üzerinden mkdir yazılarak yeni bir dosya oluşturmak gibi
```

```python
try:
  mkdir("Plots") # Plots adında directory oluştur
except: # hata kısmı-hata çıkmayacağı için boş geçtik
  pass
  mypath="Plots"
```

```python
f, axes = plt.subplots(5, 2, figsize = (15, 20)) # rows=5 cols=2 fig boyutu 15,20 olan subplot lar oluşturuldu
f.tight_layout(pad=8) # padding ayarlamak yani görseller arası boşluğu ayarlar
f.suptitle("Distribution Before Preprocessing") # sup title üst başlık
cols = df.select_dtypes(exclude="object").columns # object veri tipi işimize yaramayacağı için onu hariç tutacagız

x_axis = 0 
y_axis = 0

for col in cols: # döngü tüm col değerleri için döner ve histogram çizdirir
  sns.histplot(data=df, x=col, kde=True, ax=axes[x_axis, y_axis])
  # kde(kernel dencity estimate = çekirdek yoğunluğu tahmini)
  # bir veri kümesi üzerindeki dağılımları görselleştirirken 
  # bir veya daha fazla boyutta sürekli verimiz varsa bir olasılık yoğunluk eğrisi oluşturur
  axes[x_axis,y_axis].set_xlabel(col.title())
  axes[x_axis,y_axis].set_ylabel("Count")
  axes[x_axis,y_axis].set_title(f"{col.title()} Count")
  
  # 5 e 2 lik matriste axis değerleri matris üzerindeki koordinatları ifade eder
  # yani görselin hangi panele basılacağını belirler
  if y_axis == 1:
    y_axis = 0
    x_axis += 1
    # continue
  else:
    y_axis += 1 

plt.savefig("Plots/histograms.png")
plt.show()
```

```python
# Box Plot
f, axes = plt.subplots(5,2, figsize = (15,20))
f.tight_layout(pad=8)
f.suptitle("Boxplots")
cols = df.select_dtypes(exclude="object").columns

x_axis = 0
y_axis = 0

for col in cols:
  sns.boxplot(data=df, x=col, ax=axes[x_axis, y_axis])
  axes[x_axis,y_axis].set_xlabel(col)
  axes[x_axis,y_axis].set_ylabel("Count")
  axes[x_axis,y_axis].set_title(f"{col.title()} Box Plot")
  if y_axis == 1:
    y_axis = 0
    x_axis += 1
  else:
    y_axis += 1 

plt.savefig("Plots/boxplots.png")
plt.show()
```

```python
# Bar Plot
year_group=df.groupby("year").sum() # year kolonu için sum fonk kullanılarak gruplama yapıldı
year_group
```

```python
# Yıl bazında Positive_Affect kolonu için görselleştirme
year_group["Positive affect"].plot()
plt.savefig("Plots/Positive_Affect_Plot.png")
```

```python
# Yıl bazında PNegative_Affect kolonu için görselleştirme
p1=year_group["Negative affect"].plot()
p1.tick_params(axis="x", rotation=90) # x eksenindeki yazıların yönünü 90° döndürme
plt.savefig("Plots/Negative_Affect_Plot.png")
```

```python
ax1= sns.barplot(x=year_group.index, y=year_group['Social support'].values)
ax1.tick_params(axis="x", rotation=90) # tick_params ile x eksenindeki yazılar 90° döndürülsün
plt.savefig("Plots/SocialSupport.png")
```

```python
# HeatMap
sns.set(rc={'figure.figsize':(15,10)})
plt.title('Correlation Matrix')
sns.heatmap(df.corr(), annot=True)
plt.savefig("Plots/Correlation_matrix.png")
```

```python
# Joint Plot 
# İki değişkenli ve tek değişkenli grafiklerle iki değişkenin grafiğini çizer
sns.jointplot(data=df, x="year", y="Social support") # ikisi arasında bir korelasyon yok
sns.jointplot(data=df, x="Life Ladder", y="Social support") 
```
